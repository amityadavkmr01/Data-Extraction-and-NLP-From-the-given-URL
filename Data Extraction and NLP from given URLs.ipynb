### Creating function for removing special symbols from the text

```python
def clean_text(text):
    """
    Return clean version of the text
    """
    # Remove all non-letters and non-spaces except for hyphens and digits
    text = re.sub("[^0-9A-Za-z\- ]+", " ", text)
    # Remove all numbers except those attached to a word
    text = re.sub("(?<!\w)\d+", "", text)
    # Remove all hyphens except between two words
    text = re.sub("-(?!\w)|(?<!\w)-", "", text)
    # Remove multiple spaces and lowercase everything
    text = " ".join(text.split())
    text = text.lower()

    return text
```

### Importing stop words files that are provided

```python
StopWords_Auditor=pd.read_csv("StopWords_Auditor.txt",header=None)
StopWords_Currencies=pd.read_csv("StopWords_Currencies.txt",header=None,encoding="ISO-8859-1",sep='\n' )
StopWords_DatesandNumbers=pd.read_csv("StopWords_DatesandNumbers.txt",header=None)
StopWords_Generic=pd.read_csv("StopWords_Generic.txt",header=None)
StopWords_GenericLong=pd.read_csv("StopWords_GenericLong.txt",header=None)
StopWords_Geographic=pd.read_csv("StopWords_Geographic.txt",header=None)
StopWords_Names=pd.read_csv("StopWords_Names.txt",header=None)
```

```python
StopWords_Auditor=StopWords_Auditor[0]
StopWords_Currencies=StopWords_Currencies[0]
StopWords_DatesandNumbers=StopWords_DatesandNumbers[0]
StopWords_Generic=StopWords_Generic[0]
StopWords_GenericLong=StopWords_GenericLong[0]
StopWords_Geographic=StopWords_Geographic[0]
StopWords_Names=StopWords_Names[0]

StopWords_Auditor=list(StopWords_Auditor.values)
StopWords_Currencies=list(StopWords_Currencies.values)
StopWords_DatesandNumbers=list(StopWords_DatesandNumbers.values)
StopWords_Generic=list(StopWords_Generic.values)
StopWords_GenericLong=list(StopWords_GenericLong.values)
StopWords_Geographic=list(StopWords_Geographic.values)
StopWords_Names=list(StopWords_Names.values)

```

### Creating function for removing stop words

```python
def text_process(text):
    txt=' '.join([word for word in text.split(" ") if word.upper() not in StopWords_Auditor])
    txt1=' '.join([word for word in txt.split(" ") if word.upper() not in StopWords_Currencies])
    txt2=' '.join([word for word in txt1.split(" ") if word.upper() not in StopWords_DatesandNumbers])
    txt3=' '.join([word for word in txt2.split(" ") if word.upper() not in StopWords_Generic])
    txt4=' '.join([word for word in txt3.split(" ") if word.upper() not in StopWords_GenericLong])
    txt5=' '.join([word for word in txt4.split(" ") if word.upper() not in StopWords_Geographic])
    return ' '.join([word for word in txt5.split(" ") if word.upper() not in StopWords_Names])
```

```python
def positive_score(tokenize_text):                              # Create a function to find Positive score 
    count_pos=0
    for i in  tokenize_text:
        if i in positive:
            count_pos+=1
    return count_pos
```

```python
def negative_score(tokenize_text):                              # Create a function to find Negative score 
    count_neg=0
    for i in  tokenize_text:
        if i in negative:
            count_neg+=1
    return count_neg
```

```python
def polarity_score(count_pos,count_neg):                        # Create a function to find Polarity score 
    Polarity_Score=(count_pos-count_neg)/((count_pos+count_neg)+0.000001)
    return Polarity_Score
```

```python
def subjective_score(tokenize_text,count_pos,count_neg):        # Create a function to find Subjective score 
    Subjective_Score = (count_pos+count_neg)/((len(tokenize_text)) + 0.000001)
    return Subjective_Score
```

```python
def average_text_length_Per_Sentence(tokenize_text,no_of_sentence): # Create a function to find average text length per sentence 
    count_let=0
    for i in tokenize_text:
        for j in i:
            count_let+=1
    Average_text_Length_per_sentence = count_let / no_of_sentence
    return Average_text_Length_per_sentence
```

```python
def percentage_of_complex_words(tokenize_text):                  # Create a function to find percentage of complex words 
    vowels=['a','e','i','o','u']
    count=0
    complex_Word_Count=0
    for i in tokenize_text:
        x=re.compile('[es|ed]$')
        if x.match(i):
            count+=0
        else:
            for j in i:
                if(j in vowels):
                    count+=1
        if(count>2):
            complex_Word_Count+=1
        count=0
    Percentage_of_Complex_words = (complex_Word_Count/len(tokenize_text))*100
    return Percentage_of_Complex_words
```

```python
def fog_index(tokenize_text):                                     # Create a function to find Fog Index
    count_let=0
    for i in tokenize_text:
        for j in i:
            count_let+=1
    Avg_word_length=count_let/len(tokenize_text)
    Fog_Index=0.4 *(Avg_word_length + percentage_of_complex_words(tokenize_text))
    return Fog_Index
```

```python
def avg_num_of_words_per_sentence(tokenize_text,no_of_sentence):  # Create a function to find average number of words per sentence 
    Average_words_per_sentence = len(tokenize_text)/no_of_sentence
    return Average_words_per_sentence
```

```python
def complex_word_count(tokenize_text):                            # Create a function to find Complex word count
    vowels=['a','e','i','o','u']
    count=0
    Complex_Word_Count=0
    for i in tokenize_text:
        x=re.compile('[es|ed]$')
        if x.match(i):
            count+=0
        else:
            for j in i:
                if(j in vowels):
                    count+=1
        if(count>2):
            Complex_Word_Count+=1
        count=0
    return Complex_Word_Count
```

```python
def word_count(tokenize_text):                                    # Create a function to find word count 
    return len(tokenize_text)
```

```python
def syllable_per_word(tokenize_text):                             # Create a function to find syllable per word
    vowels=['a','e','i','o','u']
    count=0
    for i in tokenize_text:
        for j in i:
            if j in vowels:
                count+=1
    syllable_count=count/len(tokenize_text)
    return syllable_count 
```

```python
def personal_pronouns(tokenize_text):                             # Create a function to find personal pronouns
    pronouns=["i","we","my","ours","us"]
    pronouns_count=0
    for i in tokenize_text:
        if i in pronouns:
            pronouns_count+=1
    return pronouns_count
```

```python
def avg_word_length(tokenize_text):                               # Create a function to find avg word length
    count_let=0
    for i in tokenize_text:
        for j in i:
            count_let+=1
    Avg_word_length=count_let/len(tokenize_text)
    return Avg_word_length
```

### Create a empty lists 

```python
POSITIVE_SCORE=[]
NEGATIVE_SCORE=[]
POLARITY_SCORE=[]
SUBJECTIVE_SCORE=[]
AVG_TEXT_LENGTH_PER_SENTENCE=[]
PERCENTAGE_OF_COMPLEX_WORDS=[]
FOG_INDEX=[]
AVG_NUM_OF_WORDS_PER_SENTENCE=[]
COMPLEX_WORD_COUNT=[]
WORD_COUNT=[]
SYLLABLE_PER_WORD=[]
PERSONAL_PRONOUNS=[]
AVG_WORD_LENGTH=[]
```

```python
for i in text_list:
    li=[]
    a=i.split('.')            
    for j in a:
        li.append(j)
    b=pd.DataFrame()
    b['abc']=li
    b['abc']=b['abc'].apply(clean_text)
    b['abc']=b['abc'].apply(text_process)
    text_tokens=[]
    for k in b["abc"].values:
        for l in k.split(" "):
            text_tokens.append(l)
    count_pos=positive_score(text_tokens)
    POSITIVE_SCORE.append(count_pos)
    count_neg=negative_score(text_tokens)
    NEGATIVE_SCORE.append(count_neg)
    POLARITY_SCORE.append(polarity_score(count_pos,count_neg))
    SUBJECTIVE_SCORE.append(subjective_score(text_tokens,count_pos,count_neg))
    no_of_sentence=b.shape[0]
    AVG_TEXT_LENGTH_PER_SENTENCE.append(average_text_length_Per_Sentence(text_tokens,no_of_sentence))
    PERCENTAGE_OF_COMPLEX_WORDS.append(percentage_of_complex_words(text_tokens))
    FOG_INDEX.append(fog_index(text_tokens))
    AVG_NUM_OF_WORDS_PER_SENTENCE.append(avg_num_of_words_per_sentence(text_tokens,no_of_sentence))
    COMPLEX_WORD_COUNT.append(complex_word_count(text_tokens))
    WORD_COUNT.append(word_count(text_tokens))
    SYLLABLE_PER_WORD.append(syllable_per_word(text_tokens))
    PERSONAL_PRONOUNS.append(personal_pronouns(text_tokens))
    AVG_WORD_LENGTH.append(avg_word_length(text_tokens))
```
